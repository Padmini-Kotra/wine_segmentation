```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
```

```python
df=pd.read_csv("wine.csv")
```

```python
df.head()
```

```python
df.columns
```

```python
# Removing leading and trailing spaces from the column names

df.columns=df.columns.str.strip()
```

```python
df.columns
```

```python
# Removing special characters from the column names

import string
list1=list(df.columns)
col_names=[]
l=['$','@','#','?']
for i in list1:
    for j in i:
        if j in string.punctuation:
            i=i.replace(j,"")
    col_names.append(i)       
```

```python
# Replacing the column names with modified column names

df.columns=col_names
```

```python
df.rename(columns={'customersegment':'segment'},inplace=True) #renaming the name customersegment to the more appropriate segment
```

```python
df.head(50)
```

```python
# The values in 'segment' column look inconsistent. So, finding the unique values in that column

df.segment.unique()
```

```python
# Making the values in the 'segment' column consistent in the whole dataframe

df.segment = df.segment.str.upper()
df.segment.unique()
```

```python
df.segment=df.segment.map({'ONE':1,'1':1,'TWO':2,'2':2,'THREE':3,'3':3})
```

```python
# The other methods of converting categorical data to numbers are pd.Categorical, OneHotEncoding, 
# labelencoder, get_dummies()
# get_dummies will add new columns for each category
# pd.categorical will give numbers starting from 0 for every category in ascending order.
# label encoding works similar to pd.categorical
# onehotencoding works similar to get_dummies
```

```python
df.segment.unique()
```

```python
df.dtypes
```

```python
# Delete all rows with special characters if they are less in number. If their number is high, we need to find a way to impute them if possible
# Convert all data points with special characters to NaN and then drop the NaN values using dropna()

# Finding object type columns
list2=[]
for i in list(df.columns):
    if df[i].dtypes=='O':
        list2.append(i)
print(list2)
        
# Replacing the special characters in the object type columns with nan and coverting them to float type
for i in list2:
    df[i]=df[i].replace('[\$#%]',np.nan,regex=True) #regex is used for pattern searching and replacement.
    df[i]=df[i].astype(float)
df.dtypes
```

```python
df.isna().sum()
```

```python
# Dropping the nan rows as the number of null values are quite less
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
# USing zscore to find the outlier

import scipy.stats as stats
z = np.abs(stats.zscore(df))

#only keep rows in dataframe with all z-scores less than absolute value of 3 
df = df[(z<3).all(axis=1)]

#find how many rows are left in the dataframe 
df.shape
```

```python
ind_cols=df.drop('segment',axis=1)
sns.heatmap(ind_cols.corr())
plt.show()
```

```python
# Define a threshold above which we will remove all the highly correlated variables and keep only one of them

# map features to their absolute correlation values after dropping the target variable
corr = df.drop('segment', axis=1).corr().abs()

# set equality (self correlation) as zero
corr[corr == 1] = 0

# of each feature, find the max correlation
# and sort the resulting array in ascending order
corr_cols = corr.max().sort_values(ascending=False)

# display the highly correlated features
x = corr_cols[corr_cols > 0.7]
list2 = list(x.index)
print(list2)
```

```python
# From the highly correlated columns, keep one and remove others
df.drop(list2[1:],axis=1,inplace=True)
```

```python
df.head()
```

```python
# Remove duplicate entries from the dataframe

df.shape
df.drop_duplicates(inplace=True) # It will delete all the rows which are exactly the same and keep only one of them
```

```python
df.shape # check how many rows are remaining
```

```python
# Find the variance in each column and if the variance is low, we drop it and if it is high, we keep it.
# For eg, in age column if all the ages are 30, that data will not contribute to any insights of the dependent variable.

new_df=pd.DataFrame(df.var())
```

```python
new_df
```

```python
# Make a list of columns with variance less than 0.06

list1=list(new_df[new_df[0]<0.06].index)
list1
```

```python
df.drop(list1,axis=1,inplace=True) #drop the columns with very less variance
```

```python
df.head()
```

```python
# Defining X and Y (dependent and independent)
X=df.drop('segment',axis=1)
Y=df['segment']

# Scaling the data
sc=StandardScaler()
X=sc.fit_transform(X)

# Splitting the data set into train and test data
x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=5,test_size=0.25)
print("X_train shape is:",x_train.shape)
print("Y_train shape is:",y_train.shape)
print("X_test shape is:",x_test.shape)
print("Y_test shape is:",y_test.shape)
```

```python
# Ppredict using logistic regression, decision tree, random forest, Naive Bayes Classifier
# Then use Linear Discriminant Analysis(LDA) for dimensionality reduction
# Then run all the algorithms again to check accuracy.
```

```python
model_Logit=LogisticRegression()
model_Logit.fit(x_train,y_train)
pv_Logit=model_Logit.predict(x_test)
print("Accuracy score for Logistic Regression is",accuracy_score(y_test,pv_Logit))
```

```python
model_DTC=DecisionTreeClassifier()
model_DTC.fit(x_train,y_train)
pv_DTC=model_DTC.predict(x_test)
print("Accuracy score for Decision Tree is",accuracy_score(y_test,pv_DTC))
```

```python
model_RF=RandomForestClassifier(n_estimators=64)
model_RF.fit(x_train,y_train)
pv_RF=model_RF.predict(x_test)
print("Accuracy score for Random Forest is",accuracy_score(y_test,pv_RF))
```

```python
model_NB=GaussianNB()
model_NB.fit(x_train,y_train)
pv_NB=model_NB.predict(x_test)
print("Accuracy score for Naive Bayes is",accuracy_score(y_test,pv_NB))                                
```

```python
x_train.shape #check shape before LDA
x_test.shape
```

```python
#Performing Linear Discriminant Analysis and transforming the x-train and x-test accordingly

lda=LDA(n_components=2)
x_train=lda.fit_transform(x_train,y_train)
x_test=lda.transform(x_test)
print("shape of xtrain is:",x_train.shape)
print("shape of xtest is:",x_test.shape)
```

```python
model_Logit1=LogisticRegression()
model_Logit1.fit(x_train,y_train)
pv_Logit1=model_Logit1.predict(x_test)
print("Accuracy score for Logistic Regression is",accuracy_score(y_test,pv_Logit1))
```

```python
model_DTC1=DecisionTreeClassifier()
model_DTC1.fit(x_train,y_train)
pv_DTC=model_DTC1.predict(x_test)
print("Accuracy score for Decision Tree is",accuracy_score(y_test,pv_DTC))
```

```python
model_RF1=RandomForestClassifier(n_estimators=64)
model_RF1.fit(x_train,y_train)
pv_RF=model_RF1.predict(x_test)
print("Accuracy score for Random Forest is",accuracy_score(y_test,pv_RF))
```

```python
model_NB1=GaussianNB()
model_NB1.fit(x_train,y_train)
pv_NB=model_NB1.predict(x_test)
print("Accuracy score for Naive Bayes is",accuracy_score(y_test,pv_NB)) 
```

#### There is no change in accuracy after dimensionality reduction with Logistic Regression. But the accuracies with other algorithms improved.
